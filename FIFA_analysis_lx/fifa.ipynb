```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("fifa_data.csv")
data
```

### Histograms

```python
plt.figure(figsize=(7,2) , dpi=300)
bins = [40,50,60,70,80,90,100]
plt.hist(data.Overall , bins=bins)
plt.xlabel("Skill Level" , fontweight = 'bold')
plt.ylabel("Number of Players" , fontweight = 'bold')
plt.xticks(bins)
plt.show()
```

```python
left = data.loc[data['Preferred Foot']== 'Left'].count()[0]
right = data.loc[data['Preferred Foot']== 'Right'].count()[0]

labels = ['left','right']
colors = ["#008c99","#357ec7"]
plt.pie([left,right],
        labels = labels,
        shadow=True ,
        colors=colors ,
        autopct='%.2f %%',
        wedgeprops={'edgecolor' :'black','linewidth':1.5},
        textprops={'fontsize': 12, 'color': 'blue'})
plt.title("Foot Prefrence of Players" , fontdict={'fontname' : 'Comic Sans MS'})
plt.axis('equal')
plt.savefig("players_ft_preference.png")
plt.show()
```

```python
data.Weight
```

```python
data.Weight = data.Weight.str.removesuffix('lbs').astype(float)
data.Weight[0]
```

```python
light = (data.Weight<150).sum()
light
```

```python
medium = ((data.Weight >= 150) & (data.Weight < 175)).sum()
medium
```

```python
medium_heavy = ((data.Weight >= 175) & (data.Weight < 200)).sum()
medium_heavy
```

```python
heavy = (data.Weight >= 200).sum() 
heavy
```

```python
main = [light, medium ,medium_heavy , heavy]
labels = ['light', 'medium', 'medium_heavy', 'heavy']
explode = (0.2,0,0.15,0.3)
plt.figure(figsize=(12,6) , dpi=100)
plt.title("Players Weight Distribution (in lbs)" , fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 18 },pad=45 )
plt.pie(
    main,
    labels=labels,
    radius= 1.5,
    shadow=True,
    explode=explode,
    autopct='%.2f %%',
    wedgeprops={'edgecolor': 'black' , 'linewidth' : 1.5},
    textprops={'fontsize' : 12 , 'color' : 'black' , 'fontweight' :'bold'}
)
plt.tight_layout()
plt.savefig("Players_weight_dist.png",dpi = 100)
plt.show()
```

```python

barcelona  = data[data.Club == 'FC Barcelona']['Overall']
madrid  = data[data.Club == 'Real Madrid']['Overall']
ner = data[data.Club == 'New England Revolution']['Overall']
```

```python
labels = ['FC Barcelona' , 'Real Madrid' , 'NE Revolution']
plt.figure(figsize=(3.5,4) , dpi=200)
plt.title("Soccer Team Comparison", fontsize = 9 , fontdict={'fontname': 'Comic Sans MS'})
# plt.xlabel("FIFA Teams")
plt.ylabel("FIFA Ratings" , fontsize = 9)
plt.boxplot([barcelona,madrid,ner],
            tick_labels=labels)

plt.xticks(fontsize=7),
plt.yticks(fontsize=7),
plt.savefig("Fifa_team_comparison.png" ,dpi = 200)
plt.show()
```
