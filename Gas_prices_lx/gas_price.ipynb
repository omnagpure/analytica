```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("gas_prices.csv")
data.columns.values[8] = 'South_Korea'
data
```

```python
plt.figure(figsize=(11,5) , dpi=300)

plt.plot(data.Year,data.USA, color="navy", marker= 'o', label = "USA")
plt.plot(data.Year,data.South_Korea,color = 'red', marker = 'o',label="South_Korea")
plt.plot(data.Year,data.Germany,color = 'purple', marker = 'o',label = "Germany")

plt.title("Gas Prices Over Time (in $)" , fontdict={'fontname':'Comic Sans MS'} , fontweight='bold')

plt.xlabel('Year',fontweight = 'bold')
plt.ylabel('USD ',fontweight = 'bold')

plt.xticks(data.Year)
plt.yticks(np.arange(1,8,0.5))

plt.legend()

# Spines (remove top/right for cleaner look)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# # Tight layout
# plt.tight_layout()
plt.savefig("gas_price_fig.png" , dpi=300)
plt.show()
```
